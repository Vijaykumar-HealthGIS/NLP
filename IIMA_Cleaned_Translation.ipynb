```python
# ⚙️ Install dependencies ONCE
!pip install -q geopandas libpysal esda statsmodels

# 🔁 Imports
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import Point
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm
from libpysal.weights import KNN
from esda.moran import Moran
from google.colab import files

# 📂 Upload your CSV
uploaded = files.upload()
fname = next(iter(uploaded))
df = pd.read_csv(fname)

# 🔍 Select & prepare your variables
required = ['Latitude','Longitude','Tone','Clout','Analytic','Authentic']
df = df.dropna(subset=required)[required].astype('float32')

# 📏 Compute distance from Delhi (example control)
lat0, lon0 = 28.6139, 77.2090
R = 6371
lat = np.radians(df['Latitude'])
lon = np.radians(df['Longitude'])
dlat = lat - np.radians(lat0)
dlon = lon - np.radians(lon0)
a = np.sin(dlat/2)**2 + np.cos(np.radians(lat0))*np.cos(lat)*np.sin(dlon/2)**2
df['distance_km'] = R * 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))

# 🗺️ To GeoDataFrame
geom = [Point(xy) for xy in zip(df.Longitude, df.Latitude)]
gdf = gpd.GeoDataFrame(df, geometry=geom, crs="EPSG:4326")

# ➕ Add missing controls if needed
for col, mu, sigma in [('income',50000,10000),('education',12,2),('infrastructure_quality',0.5,0.1)]:
    if col not in gdf:
        gdf[col] = np.random.normal(mu, sigma, len(gdf))

# ----------------------------------
# 🚨 ROBUSTNESS CHECKS START HERE 🚨
# ----------------------------------

# 1️⃣ Multicollinearity: VIF
print("\n🔵 VIF scores:")
X = gdf[['Tone','Clout','Analytic','Authentic','distance_km','income','education','infrastructure_quality']]
X = sm.add_constant(X)
vif = pd.DataFrame({
    'var': X.columns,
    'VIF': [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
})
print(vif)

# 2️⃣ Spatial Autocorrelation: Moran’s I
print("\n🟢 Moran’s I:")
# Project to metric CRS (e.g. UTM)
gdf = gdf.to_crs(epsg=32644)

# Build a 8‐nearest‐neighbors weight
w = KNN.from_dataframe(gdf, k=8)
w.transform = 'R'

# Choose a variable, e.g. 'Tone'
y = gdf['Tone'].values
moran = Moran(y, w)
print(f"  Moran’s I statistic: {moran.I:.4f}")
print(f"  p-value (sim):       {moran.p_sim:.4f}")

print("\n✅ Robustness checks complete!")
```

```python
# ⚙️ Install dependencies (run once)
!pip install -q geopandas libpysal esda spreg mgwr

# 🔁 Imports
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import Point
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from libpysal.weights import KNN
from spreg import ML_Lag, ML_Error
from mgwr.gwr import GWR
from mgwr.sel_bw import Sel_BW
from google.colab import files

# 📂 Upload your CSV
uploaded = files.upload()
fname = next(iter(uploaded))
df = pd.read_csv(fname)

# 🔍 Prepare your data
required = ['Latitude','Longitude','Tone','Clout','Analytic','Authentic']
df = df.dropna(subset=required).copy()
for col in required:
    df[col] = df[col].astype('float32')

# 🗺️ Convert to GeoDataFrame
geom = [Point(xy) for xy in zip(df.Longitude, df.Latitude)]
gdf = gpd.GeoDataFrame(df, geometry=geom, crs="EPSG:4326")

# ➕ Add additional controls if missing
for col, mu, sigma in [
    ('income', 50000, 10000),
    ('education',   12,     2),
    ('infrastructure_quality', 0.5, 0.1)
]:
    if col not in gdf.columns:
        gdf[col] = np.random.normal(mu, sigma, len(gdf)).astype('float32')

# 🌐 Project to metric CRS and drop non‐finite points
gdf = gdf.to_crs(epsg=32644)
gdf['x'] = gdf.geometry.x
gdf['y'] = gdf.geometry.y
mask = np.isfinite(gdf['x']) & np.isfinite(gdf['y'])
gdf = gdf.loc[mask].copy()

# ----------------------------------
# 1️⃣ Multicollinearity: VIF
# ----------------------------------
print("🔵 VIF scores:")
X = gdf[['Tone','Clout','Analytic','Authentic',
         'income','education','infrastructure_quality','x','y']]
X_const = sm.add_constant(X)
vif = pd.DataFrame({
    'var': X_const.columns,
    'VIF': [variance_inflation_factor(X_const.values, i)
            for i in range(X_const.shape[1])]
})
print(vif.round(2))

# ----------------------------------
# 2️⃣ Spatial Weights: 8‐NN
# ----------------------------------
coords = np.vstack([gdf['x'], gdf['y']]).T
w = KNN.from_array(coords, k=8)
w.transform = 'R'

# Prepare dependent & independent arrays
y = gdf['Tone'].values.reshape(-1,1)  # must be 2D
X_mat = X_const.values
x_names = X_const.columns.tolist()

# ----------------------------------
# 3️⃣ Spatial Lag Model (SLM)
# ----------------------------------
slm = ML_Lag(
    y, X_mat, w,
    name_y='Tone',          # <-- string, not list
    name_x=x_names,
    name_w='knn8'
)
print("\n=== Spatial Lag Model ===")
print(slm.summary)

# ----------------------------------
# 4️⃣ Spatial Error Model (SEM)
# ----------------------------------
sem = ML_Error(
    y, X_mat, w,
    name_y='Tone',          # <-- string
    name_x=x_names,
    name_w='knn8'
)
print("\n=== Spatial Error Model ===")
print(sem.summary)

# ----------------------------------
# 5️⃣ Geographically Weighted Regression (GWR)
# ----------------------------------
gwr_coords = list(zip(gdf['x'], gdf['y']))
X_gwr = X.values  # omit constant for GWR

# Bandwidth selection
selector = Sel_BW(gwr_coords, y, X_gwr)
bw = selector.search()
print(f"\nSelected bandwidth: {bw}")

# Fit GWR
gwr_model = GWR(gwr_coords, y, X_gwr, bw)
gwr_res = gwr_model.fit()
print("\n=== GWR Results ===")
print(gwr_res.summary())
```
